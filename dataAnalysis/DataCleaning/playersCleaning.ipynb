```python
import pandas as pd
import numpy as np
import os
```

```python
file_path = '../assignmentData/players.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head())

print(df.info())

print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
missing_data_rows = df['agent_name'].isnull().sum()
total_rows = len(df)
print('Numero totale di tuple: ', total_rows)
missing_data_percentage = (missing_data_rows / total_rows) * 100
print(f'Percentuale di righe con dati mancanti: {missing_data_percentage:.2f}%')
```

```python
df = df.drop('agent_name', axis=1)
df
```

```python
print(df.duplicated().sum())
```

```python
df['country_of_citizenship'] = df['country_of_citizenship'].replace('England', 'United Kingdom')
df['country_of_citizenship'] = df['country_of_citizenship'].replace('Wales', 'United Kingdom')
df['country_of_citizenship'] = df['country_of_citizenship'].replace('Scotland', 'United Kingdom')
```

```python
cleaned_file_path = '../cleanedData/playersCleaned.csv'
if not os.path.exists(cleaned_file_path):
    df.to_csv(cleaned_file_path, index=False)
    print(f"File '{cleaned_file_path}' creato con successo.")
else:
    df.to_csv(cleaned_file_path, index=False, mode='w')
```
